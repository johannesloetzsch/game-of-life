```python
import numpy as np
import matplotlib.pyplot as plt

def grad(F):  ## vectorfield from scalarfield
    return np.array([np.gradient(F, axis=i-1) for i in range(len(F.shape))])

def div(V):  ## scalarfield of vectorfield
    div_per_dimension = np.array([grad(V[i])[i] for i in range(len(V))])
    return -np.ufunc.reduce(np.add, div_per_dimension)

c = [50,50]; ε = 50  ## location and size of center
Φ = np.fromfunction(lambda x, y: 1/np.sqrt(ε+(x-c[0])**2+(y-c[1])**2), (100, 100))  ## GravitationField
g = grad(Φ)

plt.imshow(Φ, cmap=plt.colormaps.get_cmap('YlGnBu_r')); plt.colorbar()
#plt.quiver(g[0], g[1])
plt.show()
```

```python
from functions import convolve2d
K_withAntimatter = np.array([[-1],[0],[1]])/2
K = K_withAntimatter

def plot(M, V, s):
    fig, [ax0, ax1, ax2] = plt.subplots(1,3)
    ax0.imshow(M)
    ax1.quiver(V[0], -V[1])
    #ax1.quiver(V[0][30:70,30:70], V[1][30:70,30:70])
    #ax1.quiver(V[0][40:60,40:60], V[1][40:60,40:60])
    ax1.set_aspect('equal', 'box')
    ax2.imshow(div(V), cmap=plt.colormaps.get_cmap('YlGnBu_r'))
    plt.show()

def plot_details(M, V, s):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(M)
    #ax.quiver(V[0], -V[1]); plt.show()
    plt.show()
    print("Step={:d}, Mmax={:.4f}, ||M||sum={:.4f}, Mmin={:.4f}, Vmax={:.4f}, ||V||sum={:.4f}, Vmin={:.4f}".format(s, np.max(M), np.sum(np.abs(M)), np.min(M), np.max(V), np.sum(np.abs(V)), np.min(V)))

boundaryM = "wrap"
boundaryV = "wrap"
def step(M, V, g, dt=10):
    G = convolve2d(M*V[0], K.T, mode="same", boundary=boundaryM, fillvalue=0) + \
        convolve2d(M*V[1], K, mode="same", boundary=boundaryM, fillvalue=0)
    Gv0 = convolve2d(V[0]*V[0], K.T, mode="same", boundary=boundaryV, fillvalue=0) + \
          convolve2d(V[0]*V[1], K, mode="same", boundary=boundaryV, fillvalue=0)
    Gv1 = convolve2d(V[1]*V[0], K.T, mode="same", boundary=boundaryV, fillvalue=0) + \
          convolve2d(V[1]*V[1], K, mode="same", boundary=boundaryV, fillvalue=0)
    M = M + G
    #M = M / np.max(np.abs(M))
    M = 0.5*np.tanh(M)+0.25*M
    V = V + [V[0]+1*Gv0, V[1]+1*Gv1] + g*dt
    r = 5
    for y in range(c[0]-r,c[0]+r):
        for x in range(c[1]-r,c[1]+r):
            M[y,x]=0; #V[0][y,x]=0; V[1][y,x]=0
    impulseMax = 10; V = np.divide(np.clip(V*M, -impulseMax, impulseMax), M, where=M!=0, out=V)
    #vTotalMax = np.max(np.sqrt(V[0]**2+V[1]**2))
    #V = V/np.maximum(1, vTotalMax)
    V = V/np.maximum(1, 0.5*np.max(np.abs(V)))
    V = 0.9*np.tanh(V)+0.5*V
    return M, V

print("gmax={:.4f}, gavg={:.4f}, gmin={:.4f}".format(np.max(g), np.average(g), np.min(g)))
M = 0.001*(np.random.rand(*Φ.shape)-0.5)
#M = np.zeros_like(Φ)
V = 2*np.random.rand(*([2]+list(Φ.shape)))
#V = np.zeros([2,100,100])
#M[50,35]=1; V[0][50,35]=0.0; V[1][50,35]=0.032

for s in range(1,1+10000):
    M, V = step(M, V, g)
    if s % 100 == 0:
        plot_details(M, V, s)
    elif s % 100 == 0:
        plot(M, V, s)
```
